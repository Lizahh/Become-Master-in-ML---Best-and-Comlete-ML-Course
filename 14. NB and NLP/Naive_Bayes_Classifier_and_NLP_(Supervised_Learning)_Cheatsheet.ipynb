# **✈-------------------- Bayes Theorem--------------------**

* Suppose we have three brothers A, B and C. We know what job A and C do. With the help of A and C, we can get to know that what job brother B does because B is related to A and C.

So,

Bayes theory is the **probability formula** that **uses previously known probability** to define **probabilities of related events occuring**.

## **Formula:**

**P(A|B) = [ P(B|A) . P(A) ] / P(B)**

Here,

P(A|B) = Probability of event A while event B has already happened.

P(A) = Probability of event A

P(B) = Probability of event B

P(B|A) = Probability of event B while event A has already happened.



## **Example:**

Every apartment in a building has a fire alarm detection system. However there are false alarm even with smoke instead of fire.

**Facts:**

1. Actual dangerous fire occurs 1% of the time. : P(fire) = 1/100 = 0.01

2. Fire alarms are not of every good quality. They just go off (bajnaa) 10% of the time. : P(alarm) = 1/10 = 0.1

3. When there is an actual fire, 95% of the time fire alarm goes off. : P(alarm | Fire) = 95/100 = 0.95

It means 5% of the time fire alarm doesn't go off when there is actual fire. Ugh!

**--- Question:---**

If the **fire alarm is ringining**, what is the probability that **there is actually a fire**?

**--- Solution: ---**

Given

P(A) = P(Alarm is ringining) = 1/10 = 0.1

P(B) = P(there is Fire) =  1/100 = 0.01

P(Alarm is ringing | actually a fire) = 0.95

We have to find:

If the **fire alarm is ringining**, what is the probability that **there is actually a fire**? It means alarm is already occuring, we have to find prob of fire for it. We know the even which has already occr come baad me and find prob wala pehly. So, we have to find:

P(there is fire | Alarm is ringing) = ?

Using Bayes theorem:

P(there is fire | Alarm is ringing) = P(Alarm is ringing | actually a fire) . P(there is Fire) / P(Alarm is ringining)

P(there is fire | Alarm is ringing) = [0.95 x 0.01] / 0.1 = 0.095 = 9.5%.





# **✈ Using Bayes Theorem in ML (Supervised Learning => Classification)**




* What is the purpose of creating a theorem when we can't use it in machine learning??

* We can definitely use bayes theorem into machine learning.

Suppose our dataset has n features as:

X = $x_1, x_2, x_3, x_4, ..., x_n$

Probability of belonging to class C from k number of classes is: $C_k$

As we know bayes theorem as:

P(A|B) = [ P(B|A) . P(A) ] / P(B)

So, the probability of belonging to class C **given** X features will be:

P($C_k$ | X) =[ P(X | $C_k$) . P($C_k$) ] / P(X)

Let's assume that all features are mutually independent from each other (the occurrence or outcome of one event does not provide any information or influence the probability of the occurrence or outcome of the other event).

If we make this assumption of feature independence, we can break down the joint probability P(Ck, x1, x2, x3, ..., xn) into a product of individual probabilities of each feature given the class as:

P($C_k$ | X) = P($C_k$, $x_1, x_2, x_3, x_4, ..., x_n$) / P(X)

We can apply chain rule on these all features one by one using conditional probability.

  * Suppose we have 3 classes: $C_1$, $C_2$, $C_3$

  * So, we can rewrite P($C_k$, $x_1, x_2, x_3, x_4, ..., x_n$) as:

      **1. For class 1:**
      
      P($C_1$, $x_1 | x_2, x_3, x_4, ..., x_n$)

      P($C_1$, $x_2 | x_3, x_4, x_5, ..., x_n$)

      .

      .

      .

      P($C_1$, $x_{n-1} | x_n$)

      P($C_1$, $x_n$)

      P($C_1$)

      **2. For class 2:**
      
      P($C_2$, $x_1 | x_2, x_3, x_4, ..., x_n$)

      P($C_2$, $x_2 | x_3, x_4, x_5, ..., x_n$)

      .

      .

      .

      P($C_2$, $x_{n-1} | x_n$)

      P($C_2$, $x_n$)

      P($C_2$)
       
      **3. For class 3:**
      
      P($C_3$, $x_1 | x_2, x_3, x_4, ..., x_n$)

      P($C_3$, $x_2 | x_3, x_4, x_5, ..., x_n$)

      .

      .

      .

      P($C_3$, $x_{n-1} | x_n$)

      P($C_3$, $x_n$)

      P($C_3$)

# **✈-------------------- Naive Bayes--------------------**

* Naive Bayes is the shorthand **for set of algorithms** that **use Bayes theorm** for (supervised learning) **classification**.

* Naive Bayes is commonly used for Natural Language Processing (NLP) tasks.

* But you will ask that words are not independent from each other how the assumption of naive bayes will become true in this situation??

Answer: There are alot of variations of naive bayes e.g.

1. Multinomial Naive Bayes
2. Guassian Naive Bayes
3. Complement Naive Bayes
4. Bernoulli Naive Bayes
5. Categorical Naive Bayes

Multinomial Naive Bayes is mostly used for the NLP.

## **Example:**

We have movie review dataset. People have given +ve and -ve reviews about the movies.

Someone has lebelled this dataset for us.

But, how to train on this text data?

Multinomial NB can do very well on such data by using its **count vectorization** model in which we count the frequency of words in the document.

**Step 1:**

Make classes +ve and -ve and check how many instances are there for both.

Suppose we have:

25 rows of +ve class

10 rows of -ve class

**Step 2:**

Create prior probabilities of both classes.

Total classes: 35

Rows of +ve class: 25

Rows of -ve class: 10

So,

P(+ve) = 25/35

P(-ve) = 10/35

**Step 3:**

Start count vectorization process. Count frequency of each word.

Suppose we got this:

**---->Positive Class:**

Word : Frequency

movie : 10

actor : 2

great : 8

film : 4

**---->Negative Class:**

Word : Frequency

movie : 8

actor : 10

great : 0

film : 2

So, we can see if the word great exist in review, it is positive review. If it has negative word actor in it.

**Testing:**

If we have some review as 'movie actor' or 'movie film'. Is it positive review or negative review?

Answer:

For it, let's calculate conditional probabilities of words of each class.

Total words = 10 + 2 + 8 + 4 = 24

P(movie | Positive) = 10/24 = 0.42

P(actor | Positive) = 2/24 = 0.08

P(great | Positive) = 8/24 = 0.33

P(film | Positive) = 4/24 = 0.17

Let's do the same for negative class.

Total words = 8 + 10 + 0 + 2 = 20

P(movie | Negative) = 8/20 = 0.4

P(actor | Negative) = 10/20 = 0.5

P(great | Negative) = 0/20 = 0

P(film | Negative) = 2/20 = 0.1

Now lets check the probability of 'movie actor'.

P(movie | Positive) x P(actor | Positive) x P(Positive)

= 0.42 x 0.08 x 25/35 = 0.024


P(movie | Negative) x P(actor | Negative) x P(Negative)

= 0.4 x 0.5 x 10/35 = 0.057

So, the review will belong to negative class.



















**Laplace Smoothing / Alpha Parameter:**

We can see that whenever there will be word 'great' the negative class will go to straight zero. Other attached words won't matter. So, its better to add a smoothing parameter in all words. It can be any value e.g. 1 , 2 etc.

# **Cheatsheet:**

**1. What is Naive Bayes?**

**Answer:** Naive Bayes is a classification algorithm based on Bayes' theorem with the assumption of independence among features. It is often used for text classification and spam filtering.

**2. What is Bayes' theorem?**

**Answer:** Bayes' theorem describes the probability of an event based on prior knowledge of conditions that might be related to the event. It is represented as:
P(A|B) = (P(B|A) * P(A)) / P(B)

**3. What are the assumptions made by Naive Bayes?**

**Answer:** Naive Bayes assumes that features are conditionally independent given the class label. This is known as the "naive" assumption.

**4. What are the different types of Naive Bayes algorithms?**

**Answer:** Gaussian Naive Bayes: Assumes the features follow a Gaussian distribution.
Multinomial Naive Bayes: Suitable for discrete features (e.g., word counts).
Bernoulli Naive Bayes: Suitable for binary features (e.g., presence or absence of a word).

**5. How is Naive Bayes trained?**

**Answer:** Naive Bayes is trained by estimating the prior probabilities of each class and the conditional probabilities of each feature given the class.

**6. What is Laplace smoothing (additive smoothing)?**

**Answer:** Laplace smoothing is used to handle the issue of zero probabilities. It adds a small constant value to all feature probabilities during training to prevent zero probabilities during classification.

**7. How is Naive Bayes used for text classification?**

**Answer:** In text classification, Naive Bayes can be used to determine the probability of a document belonging to a certain class based on the frequencies of words or other features in the document.

**8. What are some advantages of Naive Bayes?**

**Answer:**
Simple and fast algorithm.
Requires a small amount of training data.
Performs well in multi-class classification tasks.
Works well with high-dimensional data.

**9. What are some limitations of Naive Bayes?**

**Answer:** Naive Bayes assumes feature independence, which may not hold in some cases.
It is sensitive to irrelevant features.
Can't handle new features not seen during training.
May suffer from the "zero-frequency" problem.

**10. How do you handle continuous features with Naive Bayes?**

**Answer:** For continuous features, Gaussian Naive Bayes can be used, which assumes that features follow a Gaussian (normal) distribution.

**11. How do you handle missing data with Naive Bayes?**

**Answer:** Missing data can be handled by either ignoring the instance with missing data or using techniques like mean imputation or median imputation.

**12. What evaluation metrics can be used for Naive Bayes?**

**Answer:** Common evaluation metrics include accuracy, precision, recall, and F1 score. These metrics help assess the performance of the Naive Bayes classifier.
