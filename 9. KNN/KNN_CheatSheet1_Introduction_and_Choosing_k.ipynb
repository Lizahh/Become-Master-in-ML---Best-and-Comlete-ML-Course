# **Introduction:**

* K Nearest Neighbor is simplest classification algorithm.

* It can also be used for regression task, but its performance for that is quite poor than other algorithms.

* K-nearest neighbors (KNN) assigns a label to a data point based on the labels of its nearest neighbors in the feature space.

# **When to Use?**

* K-nearest neighbors (KNN) is commonly used when you have a small to medium-sized dataset and want to classify new data points based on the similarity to their nearest neighbors, without making strong assumptions about the underlying data distribution.

* It can be effective for both classification and regression tasks.

# **How to break this Tie??**

  * Always use odd number of K.

  * In case of tie, simply reduce K by 1 until tie is broken.

  * Randomly break tie by choosing random class from nearest points.

**How Sklearn Break this Tie?**

  * If tie happens, it will see the ordering of the neighbors. Which comes before in the data, will be selected.

So **KNN Algorithm** is:

  1. Choose the K value.

  2. Sort the features by using distance metric from the following:

      * Minkowski
      * Manhattan
      * Euclidean
      * Chebyshev

  3. Perform scaling on features. Its a must so all features come in same go.



# **Choosing best value of 'K':**

  * We want a K-value that minimizes error:

        Error = 1 - accuracy

  * There are two methods of choosing K:

    1. Elbow Method

    2. Cross Validation (CV) or Grid Search of multiple K's



# **1. Elbow Method:**

  * We train and test the model for multiple K values and then calculate errors. Then we plot them:

  image.png

  What is elbow point in it??

  Answer: Look for the point on the plot where the error rate starts to decrease more slowly. The K value at the elbow point, where the decrease in error rate slows down, is typically considered as a reasonable choice for the optimal K value.

  Like in above plot, the error rate has started reducing by less than 1% between 20-25 and rather has started increasing. So where the error starts reducing by 1%, increasing the K value only makes our system complicated. Isn't it better to stop at 20 than going to 25 and increasing complexity???

  Anyhow, elbow method is not recommended!

# **2. CV and Grid Search Method:**

  * CV and grid search takes into account the K value with the lowest error rate across multiple folds. However, it can result in a more complex model (by choosing higher value of K).

  * Consider the context of the problem to decide if larger K value is an issue.

  * **Using Pipeline in CV and Grid Search:**

  If we have some operations that we have perform on repeat, we can put them in a pipeline. So, then we only need to call that pipeline instead of calling those operations again and again.

  Its not must but convinient.
