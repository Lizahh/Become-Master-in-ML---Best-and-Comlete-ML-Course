# **Motivation:**

* PCA stands for Principal Component Analysis.

* Imagine a dataset with 30 features (also called dimensions). We will have following questions:

    1. How can we visualize those all 30 features?
    2. How we can check which features are the most important so we can ignore the less important ones? We can do this by checking correlation of features with target label but what about unlabelled data?? how to check feature importance in that case??
    3. Which features describes the most variance in the dataset? because these are the features that contribute the most to the predictive power of our model.

* This all can become possible if we reduce the 30 featurs to 2 or 3...Woah! Then we can also visualize it clearly. This is called **dimensionality reduction**.

* We can perform dimensionality reduction with PCA.

* Reduce the dimensionality with PCA then make clusters or whatever you want for classification.

# **Introduction:**

* PCA do not simply picks the most relevant features or the features with most variance.

* PCA creates new dimensional components that are **combination** of '**proportions of all features**'.

**So, covariance matrix apply linear transformation to complete original data and give us eigen vector and orthogonal eigen vectors which are our principal components.**

**3. Eigen Values:**

* Once you compute the covariance matrix (or correlation matrix) of the original data, the eigenvectors and eigenvalues are derived from that matrix.

* Each eigenvector represents a principal component. The corresponding eigenvalue associated with each eigenvector/PCA quantifies the amount of variance explained by that principal component.

* The eigenvalues are typically arranged in descending order, indicating the importance of each principal component.

* The eigenvalue of a principal component represents the amount of variance explained by that component. Higher eigenvalues indicate that the corresponding principal component captures more of the total variance in the data.

* It helps you determine how much information each principal component retains compared to the others.
