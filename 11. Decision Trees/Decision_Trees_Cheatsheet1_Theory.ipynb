# **-------------- Decision Trees Terminologies--------------**

## **1. Nodes:**

>> Nodes are the individual elements or points in a decision tree that represent a specific condition or decision.

### **Types of Nodes:**

**1. Root Node (Topmost):** The root node is the topmost node in a decision tree and serves as the starting point for making decisions. It is the first condition/feature we split on.

**2. Leaf (terminal) Node:** Leaf nodes are the final nodes in a decision tree that do not have any further branches or decisions associated with them. They represent the results. No data is going out of them.

**3. Parent Node:** A parent node is a node that has one or more child nodes connected to it. It represents a decision or condition that leads to subsequent choices. It tells the children what to do next based on certain conditions or choices.

**4. Children Nodes:** Children nodes are the nodes that are directly connected to a parent node. They represent the outcomes resulting from the decision made at the parent node.

## **2. Tree Branches (subtrees):**

>> Tree branch is a small decision tree that has a complete subtree in it. It means tree branches me paths (nodes) hoty hain. un paths ko follow kren to un k agay you reach a point where you have even more choices which lead to more paths (nodes) in the decision tree.

## **3. Pruning:**

>> Pruning is a fact that you can make our decision tree more shorter.

OR YOU CAN SAY:

>> Trimming unnecessary parts of a decision tree to simplify and improve its performance.

>> Main idea behind pruning is to avoid overfitting.



# **-------------- Gini Impurity--------------**

We have read above that we need 'information' from features. Features provide very different kind of information. We just need to check which feature provide more pure information so we can use that as top nodes.

So we can say:

**'Gini Impurity is a mathematical measurement of how 'pure' the information in dataset is.'**

We can say:

Gini Impurity ==> Class impurity



image.png

We can see here that Gini impurity is curve.

We can see that Gini impurity is zero where either the probability of class is 0 or 1:

image.png

**The formula of Gini Impurity for a class is:  p(1-p)**

Means for both classes calculate p(1-p) and then sum them both . It will be gini impurity for that dataset.

So, if we say that probability p of class using curve is: 0: 0(1-0) = 0

if we say that probability p of class using curve is: 1: 1(1-1) = 0

Seee...

We can clearly say that the curve is having highest value at 0.5 = 50% probability for both classes.

Because we are having 2 classes so: 50 + 50 = 100%.

Just keep this in mind:


**Means for both classes calculate p(1-p) and then sum them both . It will be gini impurity for that dataset.**

**p = elements of that class in dataset / total elements of dataset.**

## **Example 4: (Binary Categorical Feature)**

We have the following dataset:

image.png




We can see in it we have one feature X-URL Link. Target label is: Y-Spam

Let's check how pure this feature is to split on.

We have 2 categories in this feature: Yes or No.

**Step 1: Make tree of this feature**:

If URL is Yes, then we have 2 yes and 1 No of Spam.

If URL is No, then we have 3 no and 1 yes of Spam.

image.png

Left Leaf Gini = (2/3)(1-2/3) + (1/3)(1-1/3) = 0.44

Right Leaf Gini = (1/4)(1-1/3) + (3/4)(1-3/4) = 0.375

Gini Impurity of overall feature ==> Weighted Average of both leaves.

Left Leaf Gini = 0.44

Right Leaf Gini = 0.375

Left leaf instances = 3

Right leaf instances = 4

Total instances of feature = 7


Gini Impurity of overall feature = (3/7) x 0.44 + (4/7) x 0.375

Gini Impurity of overall feature = 0.403
