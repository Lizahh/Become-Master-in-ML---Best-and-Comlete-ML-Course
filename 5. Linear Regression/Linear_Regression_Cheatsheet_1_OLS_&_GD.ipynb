# **Introduction:**

* A Linear relationship simply means a **constant straight line relationship**.

* But in real life, we have data with alot of (features) points so its not possible to simply draw a striaght line that would touch those all points.

* One thing we can do is **draw a straight line and then minimize the distance of all points from that line.**

* The distance/difference between the real data points and the line is called residual error. This distance (residual error) can be both +ve and -ve.

* Here comes the concept of Ordinary least squares (OLS).

# HAVE ONE FEATURE: **Ordinary Least Squares (OLS):**

* Ordinary least squares (ols) works by **minimizing sum of squares of difference**/distance between the original observations/points and our predicted points. So it is this:

    Σ(y - ȳ)^2 

* Just for fun: It is ordinary thats why it is minimizing. If it'd be extra-ordinary, it would maximize definitely. Haha. just a joke!

* *IT CAN BE USED WHEN WE HAVE ONLY ONE FEATURE.*

* If we have multiple features, then we use gradient descent to scale them to minimize the distance between actual and predicted points.

* As OLS can only use one feature, so we can safely say that it gives us one straight line and we already know the equation of line is:
  
      y = mx + b

Here, 
      
      m = slope of line (ڈهلوان)

      b = intercept (a point where line crosses y-axis)

image.png

Okay now make it more simpler. We have feature x. It has data points. We want to minimize the distance between the datapoints of feature x and the line. How can we do that?

We know **mx** is a thing in line equation. It means we have m associated with x. If we will make m smaller or bigger, it will directly affect x. As smaller m will be, the more less distance there will be between line and feature x.

Okay, so now. We have to memorize two terms here. m and b. We can write both in the same format as β_1 and β_0.

So, now we can safely say that **β_0 is b and β_1 is m**. 

Also, as we know 'The distance/difference between the real data points and the line is called residual error.' So we can call this **difference as error**.

So, now we can say the line of:

    y = mx + b

can be re-written as:

ŷ = $\beta_1$ x + $\beta_0$







# **Formulas:**

Million Dollar Question:

Now, we know the equation of line in linear regression. But how we know what is the value of β_0 AND β_1 that would make us sure that those are those values that would make the distance of points and line (error) as minimum as possible values?????

Answer:

We need to calculate these two values using formulas in case of Ordinary least square means when we have one feature. If we'll be having multiple features, we will use gradient descent to find these values.

So formulas for OLS for these two terms are:

ŷ = $\beta_1$ x + $\beta_0$

$\beta_1$ = $\frac{{\sum (x - \bar{x})(y - \bar{y})}}{{\sum (x - \bar{x})^2}}$

$\beta_0$ = ȳ - $\beta_1$ x̄
 

## **Example:**



Suppose a manager wants to predict the production volume based on production hours. He has the following data:

Production Hours: 34, 35, 39, 42, 43, 47

Production Volume: 102, 109, 137, 149, 150, 158





We need to produce the production volume so we will call production volume as **Y**.

We will call the production hours as **X.**

Let's calculate the straight fitted line of it with OLS.

X = 34, 35, 39, 42, 43, 47

Y = 102, 109, 137, 149, 150, 158

$\beta_1$ = $\frac{{\sum (x - \bar{x})(y - \bar{y})}}{{\sum (x - \bar{x})^2}}$

**$\beta_1$ = 558 / 124 = 4.5 (slope of line)**

$\beta_0$ = ȳ - $\beta_1$ x̄

So, the fitting line will be: 

**ŷ = -46 + 4.5 x**

**FUTURE Prediction:**

Now, if manager wants to produce 125 production units. How many hours he should work?

We can make this prediction using line produced above:

**ŷ = -46 + 4.5 x**

**125 = -46 + 4.5 x**

x = 38

So he has to get done work of 38 hours to produce 125 production unit volume.

# **Limitation of Ordinary Least Squares:**

One example of limitation of OLS is Anscombe's Quartet.

* The Anscombe's quartet is a set of four datasets that have nearly identical simple linear regression properties, such as the same means, variances, correlation coefficients, and regression lines. 

* However, when visually inspecting the data, the relationship between the predictor and response variables is significantly different in each dataset.

* The reason it is called quartet because each of this dataset produce the same regression line for completely different 4 datasets.

So, it means we can't use OLS for all kinds of datasets.

# HAVE MULTIPLE FEATURES: **Gradient Descent**

## **Motivation:**

So, we have seen how to minimze the distance between line and feature when we have only one feature using Ordinary Least Squares.

Now suppose we have multiple features suppose while predicting price of house, we need features like area of house, number of rooms etc.

How to pass one stright line from these all features??

So, let's clear things one by one.

1. We were using only $\beta_1$ while working with one feature. 
2. So, now we have n features so we will be having $\beta_n$. Those all $\beta_n$ will be associated with n features as:
  
    ŷ = $\beta_0$ + $\beta_1$ $x_1$ + $\beta_2$ $x_2$ ... $\beta_n$ $x_n$

So, it means wooofff.. Now we have to find these all $\beta_0$, $\beta_1$, $\beta_2$ ... $\beta_n$ to find ŷ. How we are gonna memorize formulas of these all? Ugh.

GOOD NEWS! We don't have to memorize formulas of these all. Let the gradient descent do it for us ✨


## **Explanation:**

First thing first: We call the average difference between the predicted values and the actual values of the target variable as **cost function.**

Have you noted something here?

**In Ordinary Least Squares, we have one feature and we calculate 'sum of squares of difference'.**

**In Cost function, we have multiple features and we calculate average of 
'sum of square of difference'.**

By average we mean this:

  1/n Σ(y - ȳ)^2

for n features.


But!

Before reaching towards calculating cost function, we should be able to have all the beta values. Those all beta values will get multipled by X's and then we will be able to calculate predicted y. That we will compare with actual values to calculate cost function.

So it means we firstly need to calculate beta values. Then the remaining process will be easy.
Let's do this.

# **Gradient descent: Explanation**

Just keep in mind this phrase first:

    Gradient descent finds minimum β's and cost function will find errors by using them.

Gradient descent gives us the vector of all β values. ⚡
