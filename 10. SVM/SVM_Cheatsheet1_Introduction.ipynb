# **B. -------------- What is Hyperplane? --------------**

  * A hyperplane is the boundary or line that separates different classes in SVM.

  * We can call hyperplane 'a separator'.

  * In N-dim space, hyperplane is a (flat affine) subspace of N-1 dimensions.


### **Types of Hyperplane:**

1. 1D Hyperplane : Point

2. 2D Hyperplane : Line

3. 3D Hyperplane : Flat Plane

### **1. 1D Hyperplane : Point**

# **C. -------------- Maximum Margin Classifier: --------------**

* How do we know that where to draw this point, line or 2D flat plane???

image.png

Answer: We will draw the hyperplane/separator where the distance between the hyperplane and the first point of that class is maximum. This is maximum margin technique.

We apply this technique using **Maximum Margin Classifier**.

**The Maximum Margin Classifier is a method that finds the best line or boundary between different groups of data points, maximizing the distance between the line and the closest data points from each group.**

It aims to create the widest possible gap or margin between the groups, making it easier to classify new data.

## **Support Vectors:**

* When we drew the line and drew maximum margin lines, there will be some points which will be closest to the margin lines. They are helping us to decide the boundary. These are called support vectors.

* By focusing on these important data points (support vectors), the algorithm can create a reliable decision boundary, allowing it to classify new, unseen data more effectively

Bias means intentionally making the algorithm more flexible so that it can better handle the complexity and variability of the data. By adding bias, we allow the algorithm to draw a more flexible decision boundary, which can better accommodate the misclassifications and variability in the data.

>> Think of bias as giving the algorithm some extra room to make mistakes.

>>  Instead of trying to fit the training data perfectly, the algorithm (after adding bias) can create a line that can handle the noise and misclassifications better.

* Think of it as giving the algorithm some wiggle room. It can still capture the general trend and separate most of the points correctly, but it doesn't have to be too rigid and fit every single point perfectly.

* By introducing this flexibility, the algorithm becomes better at handling new, unseen data because it has learned to be more adaptable. It can generalize the patterns it learned from the training data to new situations without getting too caught up in the small details.


## **Soft Margins:**

We can add bias using different techniques.

In SVM, we use the 'Soft Margins' for this bias-variance tradeoff.

>> The soft margin allows the SVM algorithm to find a separation line or hyperplane that does the best job possible in separating the majority of points correctly while accepting a few errors. This flexibility is introduced by introducing a margin that permits some points to be on the wrong side.

>> In SVM, the soft margin allows for some misclassification and overlapping of points, introducing flexibility to the model. By allowing for these errors, the model becomes less sensitive to the training data and is more likely to generalize well to unseen data.

>> Cross-validation and regularization techniques are often used to tune the soft margin parameter and find the best trade-off.

# **Kernel Trick:**

* Kernel trick avoids recomputing the points in a higher dimensional space.

* It calculates the similarity or distance between pairs of points in this higher-dimensional space, without explicitly computing the transformation.

* It makes use of the dot product between points. We know dot product of a and b is:

     a.b = |a| |b| cos

We already know cos is used for similarity measures.

* **So, the use of kernel can be thought as a measure of similarity between original features and enlarged feature space.**
