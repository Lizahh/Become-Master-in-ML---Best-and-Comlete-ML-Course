# **Confusion Metrics and Accuracy:**

Imagine we have a dataset and we perform calculation on it. We can get four types of result:

**True Positive:** it's a correct detection of something that is actually present.

**False Positive:**  it's an incorrect detection when something is predicted to be present, but it's actually not.

**True Negative:** it's a correct recognition of something that is actually absent.

**False Negative:** it's an incorrect failure to detect something that is actually present.

**Example:**

Suppose we have a test result which shows +ve and -ve signs. That if the patient is infected, his result will have +ve sign othewise he will be having -ve sign.

# **Accuracy Paradox:**

Accuracy paradox happens when we have imbalanced dataset.

For example, we have this:

image.png

Such an imabalanced dataset!!! No case of infected was in whole dataset in Actual. Lol! But let's calculate Accuracy of this:

Accuracy = (0+95) / (0+0+95+5) = 0.95

95% accuracy??? Like seriously??? Ugh. Alarming Situation!!!

This is called accuracy paradox. It means we can't trust accuracy as performance metric for our classification model.

Here, the precision and recall comes into play.

## **3. F1-Score:**

We know the formulas of Recall and Precision as:

Recall = TP / (TP + FN)

Precision = TP / (TP + FP)

Both has same numerator. So instead of using them separately, we can combine them to make on term. That is called F1-score which is harmonic mean of precision and recall.

F1-score = $2 \times \text{Precision} \times \text{Recall} / (\text{Precision} + \text{Recall})$

**Q: Why to use Harmonic mean? Why not Normal mean?**

Answer: Harmonic mean goes to zero if either of precision or recall is zero. Which means it can better show us about the imbalanced dataset and accuracy paradox. Normal mean doesn't do this.
