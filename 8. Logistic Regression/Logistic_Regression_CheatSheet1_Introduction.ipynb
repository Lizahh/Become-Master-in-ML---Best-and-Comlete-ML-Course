# **Motivation:**

* We have two kinds of target labels or features/columns in a dataset:

  1. Continous labels : Having numerical values - No ordinal relationships
  2. Categorical labels: Having string values - Have ordinal relationships

* We used linear regression to predict a 'Continous label'.

* What about the categorical columns?? If we wanna predict them, then what??

* So, we use logistic regression to predict categorical labels. 

* So, we can say that:

  **1. Linear Regression: Regression Task**
 
  **2. Logistic Regression: Classification Task**


# **Introduction:**

* Predicting categorical labels is called **classification.**

* For example:

  Class 0: Car image

  Class 1: street image

  Class 2: Bridge image

* If we wanna make the continous labels to categorical labels, we can use discretization:

  Class 0: House price $0 - $100k
  
  Class 1: House price < $100k

* In linear regression, we calculate error metric as y - ŷ.

* In logistic regression, we can't calculate error metric. How can you do:
    car_image - street_image
  
  Lol! Its not possible. So, we will use different evaluation metric for classification i.e. accuracy, F1 score etc.

* Logistic Regression tells probability of each class too. 

# **Moving from Linear  ----✈----> Logistic Regression:**

* We can transform a linear regression to logistic regression with **logistic function/sigmoid function**:

  σ(x) = $\frac{1}{{1+e^{-x}}}$

image.png

It shows any value of X between 0 and 1.

0.5 is a cutoff for classification.

* Okay cool! 

* We know in linear regression it was β.x means β is multipled with feature X. It decides which feature is important, which not. How much importance to give to each feature. But in logistic regression??? Should we use the same? If yes, then why to use logistic function bro?????

* In logistic regression, we will multiply this β.x to σ as:

  ŷ = σ (β.x)

Okay, so we can now make following assumptions clearly, about logistic regression:

  1. We know β.x decides which feature is important and which not
  2. Now multiplying it with σ will convert all β.x values between 0 and 1 because right now, we are only talking about 2 classes: 0 and 1. Everything above 0.5 will belong to class 1 and below 0.5 means belong to class 0.

So, collectively we can say that: 

  1. Positive β --> The feature is important --> This feature will belong to a class, Can't ignore it --> So, Positive β indicates an increase in chance (likelihood) of that feature belonging to class 1. 

  2. Negative β --> The feature is not that much important --> This feature will not necessarily belong to a class --> So, negative β indicates decrease in chance (likelihood) of that feature belonging to class 1. 

  3. Bigger value of β with a feature means this feature is much stronger predictor of the likelihood of belonging to class 1. 

  4. Smaller value of β with a feature means this feature is much weaker predictor of the likelihood of (that feature) belonging to class 1. 

**Which class will have more feautres in its favour will win for that data point.**

In logistic regression, y^ is the probability of belonging to either class.

Now you must be wondering why class 1? Why not class 0?

Answer: When interpreting the beta coefficients, the focus is usually on the increase or decrease in the likelihood of belonging to the class of interest (often referred to as class 1) relative to the reference class (class 0). The positive beta coefficient suggests an increase in the likelihood of belonging to the class of interest, while a negative beta coefficient suggests a decrease in that likelihood compared to the reference class.


# **Best Fit & Maximum Likelihood:**

* Suppose we got the following fits for one dataset:

Now how to check which fit is the best fit?

Like in linear regression, we used to judge that which line is best fitting the data. The line which was having the lowest sum of squared errors was the best fit.

Now in logistic. How to check which is the best because we don't calculate RSS in it???

**The fit which have the maximum likelihood and lowest (log) loss is the best fit for that dataset in logistic regression.**

* **Likelihood**: Product of probabilities of belonging to class 1.

Let's calculate the likelihood of the following fit:

image.png

Likelihood = p (1-p)

Belonging to class 1: p = ln(0.9) x ln(0.8) x ln(0.65) x ln(0.55) = 0.006054

Belonging to class 0: 1-p = ln(1-0.3) x ln(1-0.2) x ln(1-0.08) x ln(1-0.02) = 0.000134

Likelihood for first fit = 0.006054 x 0.000134 = 8.11236e-7

**Likelihood for first fit = 8.11236e-7**

Maximum means getting closer to 1.

Same like this calculate the likelihood of other two fits too and the one having the maximum likelihood is the candidate of the best fit. We will make it 100% by calculating the log loss.

Formula of log loss:

J(x) = -1/m Σ $y^i$ log(ŷ) + (1- $y^i$ ) log(1-ŷ)

J(x) = -1/m Σ [($y^j$ log(1/ 1+$e^ΣBx$) + (1 - $y^j$) log(1 - 1/ 1+1/1+$e^ΣBx$)]
