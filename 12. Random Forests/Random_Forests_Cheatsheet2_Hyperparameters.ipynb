Random Forest has many common hyperparameters with decision tree. e.g. gini impurity etc.

But there are many parameter which are unique to random forest which are the following:

  1. n_estimators
  2. max_features
  3. bootstrap
  4. oob_score



# **1. n_estimators:**

  * It tells how many decision tree you want in your random forest.

  * More Decision trees, more opportunities to learn from a variety of features combination.

  * Random forest doesn't overfit. You can use as many DT you want. It is fast.

  * Its default value is 100.

  * Publication suggests that between 64 - 128 is good choice for decision trees. More features - More trees.

  * Unsure about choosing it? Use grid search or cross validation. Plot No. of trees vs Error (like elbow method of KNN). Where the error stops reducing, use that number.

  * If we use single tree in forest, we get high error.

  * After a certain number of trees added, two things happen:

    a. Trees becomes highly correlated and more selections doesn't reveal any more information.

    b. Adding more trees just duplicate the same information.

# **2. max_features:**

  * It tells how many features you want in your each subset.

  * Original publication suggested subset = $\log_2$(N+1) random features to choose from N number of features.

  * Current convention shows that use larger than √N number of features, given N total features. Someone also suggested that N/3 is good choice.

  * So, we can say that start with √N, grid search for other possible values till N/3.

# **3. Bootstrap:**

  * Bootrstap means 'sampling with replacement'. It means we can choose one sample more than once.

  * It means randomly selecting features and then we can even bootstrap the rows in those features. So, we will get random features with random rows.

  * For example, choose these 3 random features with random rows 2, 5, 3, 5, and 1.

  image.png

  * Why bootrapping rows? Why we can't pass all the rows?

  Answer: You can set it to False if you want. BUT! We want more randomness. More randomness means less correlation between trees. This is what our purpose is. So, we bootstrap rows too so our model generalize better on unseen data because each split is not gonna see all the rows.

  * Default value of bootstrapping is True.

# **4. Out of Bag Error (OOB Error):**

  * We know that in random forest, we bootsrap the data and calculate prediction based on all trees. We ensemble/aggregate all predictions of all trees as:

    **a. Classification:** Majority is authority - Most voted y class

    **b. Regression:** Take average of predictions of all trees

  * So, bag means:

    >> **b** --> Bootstap

    >> **ag** --> Aggregated prediction

  * Hmen pta eh bootstrapping se bht sari rows use e ni hon gi during training process (jab tree bn rye thy).

  * So hm apny model k andr e aik testing shuru kren gay. How? Jo rows select ni hui thn bootsrtapping k doraan, unko use kren gay.

  For example, hmne feature A, B , D k lie rows 2, 4, 5 choose ki thn during training. Ab isi tree ko test krny k lie hm remaining rows len gay i.e. feature A, B , D k lie rows 1, 3. So, this test will make a prediction too.

  Isi trha sb trees ki baki unselected rows k lie hm testing kr k prediction kren gay.

  Now we have two results:

    1. 5 trees : 5 prediction for training : rows 2, 5, 6 of each feature e.g.--> ŷ (for rows 2, 5, 6) **-** y (for rows 2, 5, 6)

    2. 5 trees : 5 predictions for testing :  rows 1, 3--> ŷ (for rows 1, 3) **-** y (for rows 1, 3)

image.png

  * This ŷ - y is called OOB Error or OOB Score.

  * OOB Error is hyperparameter that doesn't affect training process. It is optional. It is separate process from bootstrap.

  * Setting OOB = True : **will not** change trees

  * Setting bootstrapping = True : **will** change trees

  * OOB will not be calculated on all trees. Only on those trees it will be calculated, that didn't use all trees.

  * Its possible that all trees use all rows. So due to this uncertainty, OOB is set as False by default.
