# **Motivation:**

* We know that forest is a collection of trees. So, if we know decision tree, then why its not enough?

* **Reasons why Decision Tree is not enough:**

  1. Decision tree uses Gini Impurity, which uses some of the features and gives us final result. It doesn't uses complete data. So why I collected that remaining data? Ho skta eh us data me kch eham ho jo is ne miss krdia :/

  2. Decision tree always has the same root node because root node is the feature which has the lowest gini impurity. Ugh :/ How can a information metric (gini impurity) can decide that if gini impurity is low, then that is best feature? Ho skta eh kisi or persepective se wo feature thk na ho.

* **We can change different hyperparameters in decision tree e.g.**
  1. use some other information metric for information gain

  2. minimum gini impurity set kr den k itny low wala koi b feature root node bn skta eh

  3. Apni taraf se depth limit set kr den ta.k ziada se ziada use ho sken features.

* In sb hyperparameters ko use krny k bd b, end pe mily ga to hmen Decision tree is. Which will create those issues above mentioned kisi na kisi point pe aa k.

So what is the solution??????????????????

Solution is **RANDOM FORESTS**.

# **------------------ Random Forests - Introduction------------------**

* **Main idea behind random forest:**

    'Randomly pick features from dataset ==> Consider that as one subset.'

    'Un sb subsets ki base pe decision le lo for class prediction'
*  **Process:**

  1. Suppose we have this dataset and target label 'Y':

  image.png

  2. Choose how many features you want to pick randomly for each subset. Let's say I wanna choose 3 random features in each subset. Suppose, we have this first subset:

  image.png

  **Now the issue of having same root node in decision tree is alleviated.**

  3. Now I will only look at these features of this subset. Now I will pick one random feature from these 3 feature  as root node based on information metric (gini impurity / entropy) Don't do this thing randomly. So, now I have this tree, having only root node:

  image.png




  4. Now I have to decide by myself, if I need to split these 2 children nodes. I will not decide it randomly, I will calculate information metric and based on result, I will decide if I need to split.

  5. If I have to split based on that information, I will take another subset of 3 random features:

  image.png


  6. Fro this subset, now decide which feature is best jis ki base pe mazeed meaningful splitting ho skti eh. Decide this by calculating information metric on these 3 features. Which has lowest gini impurity will be choosen.

  image.png

  7. Now I will decide for my right white child, if I need to split that. If yes, I will choose anyother random subset for it too. I will keep on doing it, till I feel that I do not need to do anymore splitting (that if my tree is not overfitting) and all remaining are terminal nodes.

  image.png

**Choosing random features allows me to use all features of my dataset. Another drawback of decision tree covered.**

  8. Okay, so now one decision tree is buildup. But we know from our general knowledge that forest is a collection of trees. So, we will decide how many trees we want in our forest. Suppose I say I want 3 trees.

  9. Now, we have made those all 3 trees. Now we need to **ensemble** them. We will look at the prediction produced by these all 3 trees. And will ensemble them to judge the final result by following 'Majority is authority' rule, as following:

  image.png



We can call these trees as **estimators** too.

  10. Suppose, all 3 trees gave different prediction. Then take mean of their sum and then decide the class. It happens when we use continous label that is a regression task. In this case, dont count the votes rather calculate average.

  image.png





It has been proved by researchers that trees in random forest grow deeper without causing overfitting.
