# **------------------------ Hyperparameters------------------------**

DBSCAN has the following 2 key hyperparameters:

1. Epsilon

2. Minimum Number of Points

# **1. Epsilon:**

*  It sets the radius within which points are considered to be in each other's proximity (nearness).

* It is mostly set as 1.

* Choosing an appropriate epsilon value is crucial for achieving meaningful clusters in the data.

* For example, if we set epsilon = 1, we will get this size circle:

image.png

If we increase size of epsilon e.g. 2, 3 ... then the size of this circle will increase and will make clusters lesser.

If we make epsilon small enough, it will have very small circle and it will have too many unique clusters.


>> **Types of Points:**

We basically have 3 types of points in DBSCAN.

1. Core Point
2. Border Point
3. Outlier

**1. Core Point:**

* A core point is a data point that has a sufficient number of neighboring points (minimum number of points) within a specified distance (epsilon).

Suppose, DBSCAN chooses a random point. It makes circle exact around that point, by taking that point as center of circle.

Suppose we set **Minimum Number of Points = 2.**

Then it checks if it has the 2 neighborhood points. If yes, then it will be called CORE Point.

(Latest versions of DBSCAN says that include core point too in that minimum number of points. Mtlb set Minimum Number of Points by adding core point in it. So hm kahen gay Minimum Number of Points = 3. Which tells k 1 is core point and 2 minimum neighborhood points to check.)

For example, see the following figure:

image.png







In it, we can see that the point in the center of circle has two neighborhood points. Which shows that it has three points in total within epsilon distance of 1 (circle). So, we can now say that this is core point.

**2. Border Points:**

* It is just the opposite of core point. If a point in the center of the circle has less points in its neighborhood, as compared to minimum number of points we set, we will call it border point.

image.png

So, we can say that this is border point as it doesn't has 2 minimum of points in its neighbor as we decided.

Keep in mind:

Border point must have at least 1 neighborhood point. So, we can say:

 1 <= Border Point < min_points




**3. Outlier:**

Any point which is all alone. Which has not even a single point in its neighborhood, is called outlier.

It is cool that DBSCAN catches the outlier by itself, we don't need to do that by us separately through pandas.

image.png


Outlier is not included in any cluster. It is not caught by any cluster.
